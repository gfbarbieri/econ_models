Now, let's compare two segmented markets which have different demand curves and include a "world market". The world market demand curve is the sum of the segmented market demand curves.

```python
markets = [Monopolist(a=100, b=1, c=25), Monopolist(a=50, b=1/3, c=25), Monopolist(a=62.5, b=1/4, c=25)]
```

Notes:
1. The profit maximizing price in Market 1 is higher than the price in Market 2.
2. Similarly, the price elasticity of demand at the profit maximizing price in Market 1 is less elastistic (more inelastic) than the price elasticity of demand in Market 2.
3. Where no price discriminatino is possible (world market), the monopoly changes a price between the two segmented market prices at the profit maximizing point, which ultimately reduces the firms total profit.
4. The world price increases in the less reponsive market (less elastic/more inelastic) and the price decreases in the more responsive market compared to the segmented markets.
5. At the world price, total quantity produced by the firm is the same, but could go either way theoretically.

```python
for i, market in enumerate(markets):
    p, q, profit = market.profit_maximization()

    print(f"Market {i+1}: The profit maximizing price per unit is {p.evalf()}.")
    print(f"Market {i+1}: The profit maximizing quantity produced is {q.evalf()}.")
    print(f"Market {i+1}: The firm's total profit is {profit.evalf()}.")
    print(f"Market {i+1}: The price elasticity of demand at the profit maximizing price and quantity is {market.price_elasticity(p_0=p, q_0=q).simplify()}")
```
